```python
%matplotlib notebook
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from tqdm import tqdm
import sympy as sym 
import random as rand
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from scipy import integrate
from scipy.integrate import odeint
```

Punto 1

```python
def f(y,x):
    return x + 2*y 
def Solucion(x):
    return -0.5*x - (1-np.exp(2*x))/4.
N = 100
t = np.linspace(0.,4.,N)
y = 0
```

```python
yext = Solucion(t)
plt.scatter(t,yext)
```

```python
def GetAdams(f,r,t):
    
    h = t[1] - t[0]
    vx = np.zeros_like(t)
    vx[0] = r
    
    for i in range(0,4):
        actual = vx[i]
        k1 = f(actual,t[i])
        k2 = f(actual+h*k1/2,t[i]+h/2)
        k3 = f(actual+h*k2/2,t[i]+h/2)
        k4 = f(actual+h*k3,t[i]+h) 
        vx[i+1] = vx[i]+(h/6)*(k1+2*k2+2*k3+k4)  
    vxc = np.copy(vx)
    

    for i in range(5,len(t)):
        
        #Bashford
        
        presente = vx[i-1]
        pasado = vx[i-2]
        pasado_2 = vx[i-3]
        pasado_3 = vx[i-4]
        pasado_4 = vx[i-5]
        
        #valores
        k1 = f(presente, t[i-1])
        k2 = f(pasado, t[i-2])
        k3 = f( pasado_2, t[i-3])
        k4 = f(pasado_3, t[i-4])
        k5 = f(pasado_4, t[i-5])
        
        
        vx[i] = vx[i-1] + h*(1901*k1 - 2774*k2 + 2616*k3 -1274*k4+251*k5)/720 
        
        
 
        #future with my pana Multon
        
        future = vx[i]
        k6 = f(future,t[i])
        
        vxc[i] = vxc[i-1] + (h/1440)*(1427*k1 -798*k2 +482*k3 + -173*k4 +27*k5+475*k6) 
        
    
    return vx, vxc
```

```python
sol2 = GetAdams(f,y, t)

plt.scatter(t,yext, marker='x', color='y',label = "Solución Análitica")
plt.scatter(t,sol2[1],marker='.',color='k', label = "Multon")

plt.legend()
```

```python
Diferencia = abs(sol2[1]-yext)
plt.plot(t, Diferencia)
plt.xscale('log')  # Establecer escala logarítmica en el eje x
plt.yscale('log') 
```
